```python
#Establecer la conexión y verificar el server conectado, importar todas las librerías
import datetime
import pprint
from pymongo import MongoClient
uri = "mongodb+srv://test:test@topicos2-nvjms.mongodb.net/myFirstDatabase?retryWrites=true&w=majority"
client = MongoClient( uri )
print(client)
db=client.get_database("sample_supplies")
print(db)
```

```python
#Cree un cursor que verifica la cantidad total de documento
sales = db['sales']
print(sales)
num_documents = sales.count_documents({'_id' : {'$exists' : 1}})    
print ( 'Numero de documentos en la colección = ' + str(num_documents) )
```

```python
#Verifique e imprima la estructura del primer documento
pp=pprint.PrettyPrinter(indent=2)  
pp.pprint(list ( sales.find().limit(1) ))
```

# Cree un cursor que verifique las ventas en la locación de Seattle

```python
seattle = sales.find({'storeLocation':'Seattle'})
for i in seattle:
  print (i)
```

```python
#Calcule el total de documentos obtenidos en Seattle
sventa = sales.count_documents( { 'storeLocation' : 'Seattle' } )    
print ( 'Cantidad total de venta en la locación Seattle = ' + str(sventa) )
```

```python
#Venta total con nivel de satisfacción de 5 en Seattle
ssat5 = sales.count_documents( { 'storeLocation' : 'Seattle',"customer.satisfaction" : 5})
print ( 'Cantidad de venta con satisfacción a nivel 5 = ' + str(ssat5) )             
```

```python
#Total de ventas realizado por masculino y femenino
vfem=sales.count_documents( { 'storeLocation' : 'Seattle',"customer.gender" : 'F'})
vmas=sales.count_documents( { 'storeLocation' : 'Seattle',"customer.gender" : 'M'})
print ( 'Cantidad de venta por cliente femenino = ' + str(vfem) + '\nCantidad de venta por cliente masculino = ' + str(vmas) )  
```
